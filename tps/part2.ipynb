```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df_loaded = pd.read_csv('employee_salaries.csv')

print(df_loaded)
```

```python
plt.figure(figsize=(8, 6))

plt.scatter(df_loaded['Years'], df_loaded['Salary'],color='b')

plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs. Years of Experience')

plt.grid(True)
plt.show()
```

```python
x_values = df_loaded[['Years']].values  
y_values = df_loaded['Salary'].values 
print(x_values)
```

```python
model = LinearRegression()
model.fit(x_values, y_values)

years_of_experience = np.array([[10]])
predicted_salary = model.predict(years_of_experience)
print(f"Predicted salary for 10 years of experience (using Linear Regression): {predicted_salary[0]:.2f}")
```

```python
x_mean = np.mean(np.ravel(x_values).tolist())
y_mean = np.mean(np.ravel(y_values).tolist())

print(x_mean, y_mean)

numerator = np.sum((np.ravel(x_values).tolist() - x_mean) * (np.ravel(y_values).tolist() - y_mean))
denominator = np.sum((np.ravel(x_values).tolist() - x_mean)**2)
beta_1 = numerator / denominator
beta_0 = y_mean - beta_1 * x_mean
print(f"β₀ (y-intercept): {beta_0:.4f}")
print(f"β₁ (slope): {beta_1:.4f}")
```

```python
plt.figure(figsize=(8, 6))
y_pred = beta_0 + beta_1 * x_values

plt.scatter(df_loaded['Years'], df_loaded['Salary'],color='b')
plt.plot(x_values, y_pred, color='red', label='Line of best fit')

plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs. Years of Experience')

plt.grid(True)
plt.show()
```

```python
predicted_salary_for_10_year = beta_0 + beta_1 * 10
print(f"Predicted salary for 10 years of experience: {predicted_salary_for_10_year:.2f}")
```
