```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Create an array containing numbers from 1 to 9
arr = np.arange(1, 10)
print('array :' ,arr)
```

```python
reshaped_arr = arr.reshape(3, 3)
print(f'Reshaped 2D array of Dimension {reshaped_arr.shape} :\n {reshaped_arr}')
```

```python
first_row = reshaped_arr[0, :]
first_column = reshaped_arr[:, 0]
print(f'first row is {first_row} and first colum is {first_column}')
```

```python
arr2 = np.arange(9, 0, -1).reshape(3, 3)
matrix_product = np.dot(reshaped_arr, arr2)

print(f'Matrix product between arr1 and arr2 \n {matrix_product}')
```

```python
X = [-1, 0, 1, 2]
Y = [3, 2, 4, 1]

# plt.plot(X, Y, marker='o')
# plt.plot(X, Y, 'r.')
plt.plot(X, Y)


plt.xlabel("X (Abscissas)")
plt.ylabel("Y (Ordinates)")
plt.title("Plot of X vs Y")

plt.grid(True) 
plt.show()
```
